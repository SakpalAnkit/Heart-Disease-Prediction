Visualizations for Understand The Data

```python
# Count values for heart disease
heart_counts = Heart_disease["target"].value_counts()

# Bar chart
plt.figure(figsize=(6,4))
plt.bar(heart_counts.index, heart_counts.values, color=['blue', 'red'])
plt.xticks([0, 1],["No Disease", "Disease"])
plt.xlabel("Heart Disease")
plt.ylabel("Count")
plt.title("Heart Disease Distribution")
plt.show()
```

```python
#This Countplot represents the Heart disease by gender
plt.figure(figsize=(6, 4))
sns.countplot(x="sex",hue="target",data=Heart_disease,palette={"No": "yellow", "Yes": "purple"})  # 0 = No Disease (blue), 1 = Disease (red)
plt.title("Heart Disease by Gender")
plt.xlabel("Gender (0 = Female, 1 = Male)")
plt.ylabel("Count")
plt.legend(title="Heart Disease", labels=["No Disease", "Disease"])
plt.xticks()
plt.show()
```

```python
#This histplot represents the Age distribution
plt.figure(figsize=(8,5))
sns.histplot(Heart_disease["age"], bins=20, kde=True, color="blue")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7, 5))
sns.countplot(x="cp", hue="target", data=Heart_disease, palette=['orange', 'green'])

plt.title("Chest Pain Type vs Heart Disease")
plt.xlabel("Chest Pain Type (cp)")
plt.ylabel("Count")
plt.legend(title="Heart Disease", labels=["No Disease", "Disease"])
plt.tight_layout()
plt.show()
```

```python
#This plot is represent the restecg type vs Heart Disease
restecg_counts = pd.crosstab(Heart_disease['restecg'], Heart_disease['target'])

restecg_counts.plot(kind='bar', stacked=True, figsize=(7,5), color=["#67b3ff", "#ff7866"])
plt.title("Heart Disease by Resting ECG Type (Stacked)")
plt.xlabel("RestECG Type")
plt.ylabel("Count")
plt.legend(["No Disease", "Disease"], title="Heart Disease")
plt.show()
```

```python
#This plot is represent the exercise induced angina (exang) type vs Heart Disease
plt.figure(figsize=(7,5))
sns.countplot(x="exercise induced angina (exang)", hue="target", data=Heart_disease, palette="coolwarm")
plt.title("exercise induced angina (exang) Type vs Heart Disease")
plt.xlabel("exercise induced angina (exang) Type")
plt.ylabel("Count")
plt.legend(["No Disease", "Disease"])
plt.show()
```

```python
#This plot is represent the Slop type vs Heart Disease
# Group by 'slope' and 'target', then reshape for plotting
slope_counts = Heart_disease.groupby(['slope', 'target']).size().unstack()

# Create subplots: one pie chart per slope type
fig, axes = plt.subplots(1, 3, figsize=(18, 6))  # Assuming there are 3 slope types: 0, 1, 2

colors = ['#66b3ff', '#ff6666']  # Colors for No Disease and Disease

for i, slope_type in enumerate(slope_counts.index):
    axes[i].pie(
        slope_counts.loc[slope_type],
        labels=['No Disease', 'Disease'],
        autopct='%1.1f%%',
        colors=colors,
        startangle=140
    )
    axes[i].set_title(f'Slope = {slope_type}')

plt.suptitle('Slope Type vs Heart Disease', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#This plot is represent the CA type vs Heart Disease
plt.figure(figsize=(10,5))
sns.countplot(x="ca", hue="target", data=Heart_disease, palette="coolwarm")
plt.title("ca Type vs Heart Disease")
plt.xlabel("ca Type")
plt.ylabel("Count")
plt.legend(["No Disease", "Disease"])
plt.show()
```

```python
#This plot is represent the Max Heart Rate (thalach)
plt.figure(figsize=(8,5))
sns.scatterplot(x="age", y="thalach", hue="target", data=Heart_disease, palette="coolwarm")
plt.title("Max Heart Rate vs. Age")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate (thalach)")
plt.legend()
plt.show()
```

Encoding 

```python
#Encoding The Object Data
from sklearn.preprocessing import LabelEncoder
Heart_disease.head()
```

```python
#Check the Data Types
Heart_disease.dtypes
```

```python
Heart_disease['cp'].unique()
```

Encode Gender

```python
encoder_sex=LabelEncoder()
encoder_sex_feature=encoder_sex.fit_transform(Heart_disease['sex'])
encoded_df=pd.DataFrame(encoder_sex_feature,columns=['encoded_sex'])
```

```python
Heart_disease.drop(['sex'],axis=1,inplace=True)
```

```python
Heart_disease=pd.concat([Heart_disease,encoded_df],axis=1)
```

Encode CP(chest pain)

```python
encoder_cp=LabelEncoder()
encoder_cp_feature=encoder_cp.fit_transform(Heart_disease['cp'])
encoded_df=pd.DataFrame(encoder_cp_feature,columns=['encoded_cp'])
```

```python
Heart_disease.drop(['cp'],axis=1,inplace=True)
```

```python
Heart_disease=pd.concat([Heart_disease,encoded_df],axis=1)
```

Encode fbs(Fasting Blood Sugar)

```python
encoder_fbs=LabelEncoder()
encoder_fbs_feature=encoder_fbs.fit_transform(Heart_disease['fbs'])
encoded_df=pd.DataFrame(encoder_fbs_feature,columns=['encoded_fbs'])
```

```python
Heart_disease.drop(['fbs'],axis=1,inplace=True)
```

```python
Heart_disease=pd.concat([Heart_disease,encoded_df],axis=1)
```

Encode restecg(Resting Electrocardiographic Result)

```python
encoder_restecg=LabelEncoder()
encoder_restecg_feature=encoder_restecg.fit_transform(Heart_disease['restecg'])
encoded_df=pd.DataFrame(encoder_restecg_feature,columns=['encoded_restecg'])
```

```python
Heart_disease.drop(['restecg'],axis=1,inplace=True)
```

```python
Heart_disease=pd.concat([Heart_disease,encoded_df],axis=1)
```

Encode exercise induced angina (exang)

```python
encoder_exang=LabelEncoder()
encoder_exang_feature=encoder_exang.fit_transform(Heart_disease['exercise induced angina (exang)'])
encoded_df=pd.DataFrame(encoder_exang_feature,columns=['encoded_exang'])
```

```python
Heart_disease.drop(['exercise induced angina (exang)'],axis=1,inplace=True)
```

```python
Heart_disease=pd.concat([Heart_disease,encoded_df],axis=1)
```

Encode slope

```python
encoder_slope=LabelEncoder()
encoder_slope_feature=encoder_slope.fit_transform(Heart_disease['slope'])
encoded_df=pd.DataFrame(encoder_slope_feature,columns=['encoded_slope'])
```

```python
Heart_disease.drop(['slope'],axis=1,inplace=True)
```

Encode The Final Result(target)

```python
encoder_target=LabelEncoder()
encoder_target_feature=encoder_target.fit_transform(Heart_disease['target'])
encoded_df=pd.DataFrame(encoder_target_feature,columns=['encoded_target'])
```

Group the X Features and Y Features

```python
x=Heart_disease[['age','trestbps','chol','thalach','oldpeak','encoded_sex','encoded_cp','encoded_fbs','encoded_restecg','encoded_exang','encoded_slope','encoded_ca','encoded_thal']]
y=Heart_disease['encoded_target']
```

Train Test And Split the Data 

Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
Ls_model=LogisticRegression()
Ls_model.fit(x_train,y_train)
```

```python
y_pred = Ls_model.predict(x_test)
```

```python
Heart_disease1 = pd.DataFrame({'Actual':y_test.values,'Predicted':y_pred})
Heart_disease1.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns    #Provides advanced statistical visualizations.

plt.figure(figsize=(8, 5))  #Creates a blank canvas with a size of 8 inches wide and 5 inches tall.

# Scatter plot for training data
plt.scatter(y_train, range(len(y_train)), color='blue', label="Train Data", alpha=0.6)  #alpha=0.6 → Adjusts transparency (1 = solid, 0 = transparent).

# Scatter plot for testing data
plt.scatter(y_test, range(len(y_test)), color='red', label="Test Data", alpha=0.6)

plt.xlabel("Target (y)")
plt.ylabel("Index")
plt.title("Scatter Plot of Train vs. Test Data")
plt.legend()   #Displays the blue (train) and red (test) labels.
plt.show()
```
